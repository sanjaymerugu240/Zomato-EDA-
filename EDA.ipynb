```python


df_country = pd.read_excel(r"C:\Users\sanja\Downloads\Zomatodataset\Country-Code.xlsx")
print(df.head())
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
# checking the datatypes
print(final_df.info())
```

```python
# knowing the count of each country
Country_names=final_df.Country.value_counts().index
```

```python
Country_val=final_df.Country.value_counts().values
```

```python
## pie plot.  Top 3 Countries that uses zomato 
plt.pie(Country_val[:3],labels=Country_names[:3],autopct= '%1.2f%%')
plt.show()
```

Observation: Zomato records or transaction are from india
After that USA and then United kingdoms

```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
Ratings
```

## observations
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when rating is between 4.0 to 4.4 ---> very good
3. when rating is between 3.5 to 3.9 ----> good
4. when ratind is between 3.0 to 3.4 ----> average 
5. when rating is between 2.0 to 2.9=-----> average
6. when Rating is between 0.0 to 2.8 ----> poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x= "Aggregate rating",y ="Rating count",data = Ratings)
plt.show()
```

```python
sns.barplot(x= "Aggregate rating",y ="Rating count",hue = 'Rating color',data = Ratings)
plt.show()
```

## Observation
1. NOt rating count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x= 'Rating color',hue = 'Rating color',data =Ratings)
plt.show()
```
